# Clustering: Assignment Questions

## Theoretical Questions

### 1. What is unsupervised learning in the context of machine learning?

**Unsupervised learning** is a type of machine learning where the algorithm is given data that has not been labeled or categorized. Instead of being told what to look for, the algorithm must learn to identify patterns, structures, and relationships within the data on its own. The two main tasks in unsupervised learning are **clustering** (grouping similar data points together) and **dimensionality reduction** (reducing the number of variables).

### 2. How does K-Means clustering algorithm work?

The K-Means clustering algorithm works by iteratively partitioning a dataset into a predefined number of clusters, 'K'. The steps are as follows:

1.  **Choose K:** The user specifies the number of clusters (K) to be found.
2.  **Initialize Centroids:** The algorithm randomly selects K data points from the dataset to serve as the initial cluster centers (centroids).
3.  **Assign Points to Clusters:** Each data point is assigned to the nearest centroid, typically based on Euclidean distance. This forms K initial clusters.
4.  **Recalculate Centroids:** The centroid of each cluster is recalculated by taking the mean of all the data points assigned to that cluster.
5.  **Repeat:** Steps 3 and 4 are repeated until the cluster assignments no longer change or a maximum number of iterations is reached. At this point, the algorithm has converged.

### 3. Explain the concept of a dendrogram in hierarchical clustering.

A **dendrogram** is a tree-like diagram used to visualize the arrangement of the clusters produced by hierarchical clustering. 

- **Structure:** The leaves of the tree are the individual data points. As you move up the tree, branches merge, representing the fusion of clusters.
- **Height:** The vertical axis of the dendrogram represents the distance or dissimilarity between clusters. The height of the horizontal line where two clusters merge indicates the distance at which this merge occurred.
- **Interpretation:** By cutting the dendrogram at a certain height, you can determine the number of clusters at that level of similarity. A longer vertical line indicates that two clusters are more distinct from each other.

### 4. What is the main difference between K-Means and Hierarchical Clustering?

The main difference lies in their approach and output:

- **K-Means Clustering** is a **partitional** (or flat) clustering algorithm. It divides the data into a pre-specified number of non-overlapping clusters (K). The final output is a single set of clusters.
- **Hierarchical Clustering** creates a **hierarchy** of clusters. It does not require the number of clusters to be specified beforehand. The output is a tree-like structure (a dendrogram) that shows the nested grouping of data points at all levels of granularity.

### 5. What are the advantages of DBSCAN over K-Means?

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) has several key advantages over K-Means:

1.  **Arbitrarily Shaped Clusters:** K-Means assumes that clusters are spherical. DBSCAN can find clusters of any shape (e.g., elongated, concentric) because it groups points based on density.
2.  **No Need to Specify K:** DBSCAN does not require the user to specify the number of clusters beforehand. It determines the number of clusters based on the data's density.
3.  **Outlier Detection:** DBSCAN has a built-in mechanism for identifying noise points (outliers) that do not belong to any cluster.

### 6. When would you use Silhouette Score in clustering?

The **Silhouette Score** is used to evaluate the quality of the clusters created by an algorithm. You would use it in the following scenarios:

- **Finding the Optimal K:** It is particularly useful for finding the optimal number of clusters (K) for algorithms like K-Means. You can calculate the Silhouette Score for a range of K values, and the value that gives the highest score is often considered the best choice.
- **Comparing Different Models:** It can be used to compare the performance of different clustering algorithms on the same dataset.

The score measures how well-separated the clusters are by calculating both the mean intra-cluster distance and the mean nearest-cluster distance for each sample.

### 7. What are the limitations of Hierarchical Clustering?

Hierarchical Clustering has some notable limitations:

- **Computational Complexity:** The time complexity is typically O(nÂ³), which makes it very slow and memory-intensive for large datasets.
- **Greedy Approach:** Once a merge or split is made, it cannot be undone. This means an early, suboptimal decision can propagate through the rest of the clustering process.
- **Difficulty with Interpretation:** While dendrograms are visually appealing, interpreting them to choose the right number of clusters can be subjective and difficult.

### 8. Why is feature scaling important in clustering algorithms like K-Means?

Feature scaling is crucial for distance-based algorithms like K-Means because they use measures like Euclidean distance to determine the similarity between data points. If the features in the dataset are on different scales (e.g., age in years and income in dollars), the feature with the larger range will completely dominate the distance calculation. This will cause the algorithm to be biased towards that feature. **Scaling** (such as Standardization or Normalization) brings all features to a comparable scale, ensuring that each one contributes equally to the distance computation.

### 9. How does DBSCAN identify noise points?

DBSCAN identifies noise points based on its concept of density. It classifies points into three types:

- **Core Point:** A point that has at least `min_samples` (a user-defined parameter) within its `eps` radius.
- **Border Point:** A point that is within the `eps` radius of a core point but does not have enough neighbors to be a core point itself.
- **Noise Point:** A point that is neither a core point nor a border point. It is not in a dense region and does not belong to any cluster.

Any point that doesn't meet the criteria to be a core or border point is labeled as noise (typically with the label -1).

### 10. Define inertia in the context of K-Means.

**Inertia** in K-Means is the **sum of squared distances of samples to their closest cluster center**. It is a measure of how internally coherent the clusters are. A lower inertia value means that the clusters are more dense and well-defined. However, inertia always decreases as the number of clusters (K) increases, so it cannot be used alone to determine the optimal K.

### 11. What is the elbow method in K-Means clustering?

The **Elbow Method** is a heuristic used to find the optimal number of clusters (K) in K-Means. It involves running the K-Means algorithm for a range of K values and plotting the inertia for each K. As K increases, inertia decreases. The plot typically looks like an arm. The point where the rate of decrease sharply changes (the "elbow" of the arm) is considered a good estimate for the optimal number of clusters, as adding more clusters beyond this point provides diminishing returns.

### 12. Describe the concept of "density" in DBSCAN.

In DBSCAN, "density" is defined by two key parameters:

1.  **`eps` (epsilon):** A distance value that defines a radius around each data point. This radius is called the point's neighborhood.
2.  **`min_samples`:** The minimum number of data points that must be within a point's `eps` neighborhood (including the point itself) for that region to be considered "dense."

A dense region is formed around a **core point**, which is a point that satisfies the `min_samples` condition. Clusters are then formed by connecting nearby core points and their neighbors.

### 13. Can hierarchical clustering be used on categorical data?

Yes, hierarchical clustering can be used on categorical data, but it requires an appropriate distance metric. Standard metrics like Euclidean distance are designed for numerical data. To handle categorical or mixed data types, you would need to use a different metric, such as:

- **Gower's Distance:** A metric specifically designed to handle mixed data types (numerical, categorical, binary).
- **Jaccard Distance:** Used for binary data (sets).

### 14. What does a negative Silhouette Score indicate?

A **negative Silhouette Score** for a data point indicates that it has been **assigned to the wrong cluster**. It means that the average distance from that point to the points in a neighboring cluster is smaller than its average distance to the points in its own assigned cluster. Essentially, the point is a better fit for another cluster.

### 15. Explain the term "linkage criteria" in hierarchical clustering.

**Linkage criteria** determines how the distance between two clusters is calculated. This is fundamental to the merging process in agglomerative hierarchical clustering. Common types of linkage include:

- **Single Linkage:** The distance between two clusters is the minimum distance between any single data point in the first cluster and any single data point in the second cluster.
- **Complete Linkage:** The distance is the maximum distance between any single data point in the first cluster and any single data point in the second cluster.
- **Average Linkage:** The distance is the average distance between all pairs of data points between the two clusters.
- **Ward's Linkage:** This method merges the two clusters that result in the minimum increase in the total within-cluster variance.

### 16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

K-Means performs poorly on such data because of its underlying assumptions. It inherently assumes that clusters are **spherical, have similar sizes, and similar densities**. 

- **Varying Sizes:** K-Means tends to create clusters of roughly equal size because its objective is to minimize inertia, which can be achieved by splitting large clusters.
- **Varying Densities:** It struggles with clusters of different densities because the centroid-based approach may place the centroid of a sparse cluster closer to a dense one, leading to misclassification of points at the boundary.

### 17. What are the core parameters in DBSCAN, and how do they influence clustering?

The two core parameters in DBSCAN are `eps` and `min_samples`.

- **`eps` (epsilon):** This is a distance value that defines the radius of the neighborhood around a data point. 
  - **Influence:** A smaller `eps` value will result in more, smaller, and denser clusters, as fewer points will be considered neighbors. A larger `eps` will merge more points into fewer, larger clusters.

- **`min_samples`:** This is the minimum number of data points required to form a dense region (i.e., to define a core point).
  - **Influence:** A higher `min_samples` value requires clusters to be denser and is more likely to label points in sparse regions as noise. A lower value will allow sparser regions to form clusters.

### 18. How does K-Means++ improve upon standard K-Means initialization?

K-Means++ is a smart initialization technique that improves upon the random initialization of standard K-Means. While standard K-Means picks centroids randomly, which can lead to poor clustering and slow convergence, K-Means++ initializes centroids to be far apart from each other.

The process is:
1.  The first centroid is chosen randomly.
2.  Each subsequent centroid is chosen from the remaining data points with a probability proportional to the squared distance from the point's closest existing centroid.

This approach leads to **better and more consistent final clusters** and often **faster convergence**.

### 19. What is agglomerative clustering?

**Agglomerative clustering** is the most common type of hierarchical clustering. It follows a **"bottom-up"** approach:

1.  **Start:** Each data point is treated as its own individual cluster.
2.  **Iterate:** In each step, the two closest clusters are merged into a single new cluster based on the chosen linkage criterion.
3.  **End:** This process is repeated until only one cluster remains, containing all the data points.

The entire sequence of merges is represented by a dendrogram.

### 20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Silhouette Score is a better metric because it considers both **cohesion** (how close points are within a cluster) and **separation** (how far apart different clusters are).

- **Inertia** only measures cohesion. Its value will always decrease as the number of clusters (K) increases, making it difficult to find the optimal K. A model with K=n (where n is the number of data points) would have an inertia of 0 but would be useless.
- **Silhouette Score** provides a more balanced view. A high score requires that points are close to others in their cluster and far from points in other clusters. This helps identify a K that provides a meaningful balance between the number of clusters and their quality.

### 21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

kmeans = KMeans(n_clusters=4, random_state=0, n_init='auto')
y_kmeans = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.title('K-Means Clustering on Synthetic Blobs')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```

### 29. Create synthetic data using make_blobs and evaluate silhouette scores for k=2 to 5. Display as a bar chart.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

k_range = range(2, 6)
silhouette_scores = []
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

plt.figure(figsize=(8, 5))
plt.bar(k_range, silhouette_scores)
plt.title('Silhouette Scores for Different Values of K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.show()
```

### 30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import dendrogram, linkage

iris = load_iris()
X = iris.data

linked = linkage(X, method='average')

plt.figure(figsize=(12, 8))
dendrogram(linked,
            orientation='top',
            labels=None,
            distance_sort='descending',
            show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

### 34. Load the Breast Cancer dataset and compare inertia values for K=2 to 6 using K-Means. Show results in a line plot.

### 44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples

X, y = load_iris(return_X_y=True)

kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
cluster_labels = kmeans.fit_predict(X)

sample_silhouette_values = silhouette_samples(X, cluster_labels)

plt.figure(figsize=(10, 6))
y_lower = 10
for i in range(3):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i
    plt.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_silhouette_values)
    y_lower = y_upper + 10

plt.title("Silhouette Plot for Each Sample")
plt.xlabel("Silhouette coefficient values")
plt.ylabel("Cluster label")
plt.show()
```

### 46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

### 47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

```python
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN

X, y = make_blobs(n_samples=300, centers=3, cluster_std=1.2, random_state=42)

dbscan = DBSCAN(eps=0.7, min_samples=5)
labels = dbscan.fit_predict(X)

n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_noise_ = np.sum(labels == -1)

print(f"Number of clusters found: {n_clusters_}")
print(f"Number of noise points found: {n_noise_}")
```

### 48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = load_digits()
X = digits.data

perplexity_value = min(30, len(X) - 1)
X_tsne = TSNE(n_components=2, random_state=42, perplexity=perplexity_value).fit_transform(X)

agg_cluster = AgglomerativeClustering(n_clusters=10)
labels = agg_cluster.fit_predict(X_tsne)

plt.figure(figsize=(10, 8))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='viridis', s=20)
plt.title('Agglomerative Clustering on t-SNE-reduced Digits Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Cluster ID')
plt.show()
```
